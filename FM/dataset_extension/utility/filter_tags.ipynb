```python
import pandas as pd
import numpy as np
import pickle
from tqdm.notebook import tqdm

tags = pd.read_pickle('../data/us/dict.pkl')
tags
```

```python
tags_keep = {
    'electronic':0,
    'rock':1,
    'indie':2,
    'jazz':3,
    'reggae':4,
    'british':5,
    'punk':6,
    '80s':7,
    'dance':8,
    'acoustic':9,
    'rnb':10,
    'hardcore':11,
    'alternative':12,
    'blues':13,
    'country':14,
    'Classical':15,
    'rap':16,
    'metal':17,
    'Hip-Hop':18,
    'hip hop':18
}
```

```python
# Top 5 tags
tags_encoded = {}
for item in tqdm(tags.items()):
    v = []
    hiphop = False
    for tag in item[1]:
        if tag in tags_keep:
            if tag == 'hip hop' or tag == 'Hip-Hop':
                if not hiphop:
                    hiphop = True
                else:
                    continue
            v.append(tags_keep[tag])
                
    if len(v) > 0:
        tags_encoded[item[0]] = v
```

```python
tags_encoded
```

```python
# Only one tag
tags_encoded = {}
for item in tqdm(tags.items()):
    t = -1
    for tag in item[1]:
        if tag in tags_keep:
            t = tags_keep[tag]
            break
                
    if t > -1:
        tags_encoded[item[0]] = t
```

```python
count = 0
for item in tqdm(tags.items()):
    for tag in item[1]:
        if tag in tags_keep:
            count += 1
            break
count
```
