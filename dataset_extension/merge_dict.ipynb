```python
d = pd.read_pickle('./data/dict_us.pkl')
```

```python
len(d)
```

```python
d_v = list(d.values())
```

```python
d_v[:5]
```

```python
count = {}
for tags in tqdm(d_v):
    for tag in tags:
        if tag in count:
            count[tag] += 1
        else:
            count[tag] = 1
```

```python
count
```

```python
count_sotred = sorted(count.items(), key=lambda x: x[1], reverse=True)
count_sotred
```

```python
len(count_sotred)
```

```python
import pickle
pickle.dump(count_sotred, open("dict_counts.pkl", "wb")) 
```
